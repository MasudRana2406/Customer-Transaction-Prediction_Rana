5.2 Handling Duplicate value

5.3 Handling Missing Value

```python
pd.set_option('display.max_rows',None) # to see all features summation of null values
data.isnull().sum()

# No missing values
```

5.4 Checking categorical variables & convert to numerical

5.5 Checking outliers & handling outliers

```python
plt.figure(figsize=(20,25),facecolor='white')
plotnumber = 1

for i in data.drop('target',axis=1):
  if plotnumber<200:
    plt.subplot(20,10,plotnumber)
    sns.boxplot(data[i],orient='h')
    plt.xlabel(i)
    plotnumber+=1
plt.tight_layout()

# No feature significant outliers to remove
```

### Feature Selection

---



```python
x = data.drop('target',axis=1)
y = data.target
```

### Dimensionality reduction

```python
from sklearn.decomposition import PCA

pca = PCA(n_components='mle',svd_solver='full')

X = pca.fit_transform(x)
X
```

### Train & test split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=73)
```

5.6 Scaling down data

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

5.7 Transformation to normal distribution

```python
# Check if data are normally distributed
plt.figure(figsize=(20,25),facecolor='white')
plotnumber = 1

for i in data.drop('target',axis=1):
  if plotnumber<200:
    plt.subplot(20,10,plotnumber)
    sns.histplot(x=i,data=data)
    plt.xlabel(i)
    plotnumber+=1
plt.tight_layout()

# All data are normally distributed
```

5.8 Handling Imbalanced dataset

```python
# check if data set imbalanced
(data['target'].value_counts()/data['target'].value_counts().sum())*100
```

```python
plt.pie(data['target'].value_counts(),labels=data['target'].unique(),autopct='%0.2f%%')
```

- Found data is imbalance

```python
# Using SMOTE to balance dataset
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_bal,y_bal = smote.fit_resample(X_train_scaled,y_train)
```

```python
plt.pie(y_bal.value_counts(),labels=y_bal.unique(),autopct='%0.2f%%')
```

```python
# Check if now balanced using Counter
from collections import Counter

print(Counter(y))
print(Counter(y_bal))
```

## Hyperparameter Tuning

```python
param_dist = {
    'penalty': ['l1', 'l2'],
    'C': [0.1, 1.0, 10.0],
    'fit_intercept': [True, False],
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 500]
}

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

modelLR = LogisticRegression()

tuned_modelLR = GridSearchCV(modelLR,param_dist,refit = True, verbose = 2,scoring = 'accuracy',cv=5)
```

```python
tuned_modelLR.fit(X_bal,y_bal)
```

```python
print("Tuned SVM Parameters: {}".format(tuned_modelLR.best_params_))
print("Best score is {}".format(tuned_modelLR.best_score_))
```

**Logistic Regression tuned model with best parameter**

```python
from sklearn.linear_model import LogisticRegression

Best_model = LogisticRegression(penalty='l2',C=0.1,fit_intercept=True,solver='liblinear',max_iter=100)

Best_model.fit(X_bal,y_bal)

y_predict = Best_model.predict(X_test_scaled)

y_predict
```
